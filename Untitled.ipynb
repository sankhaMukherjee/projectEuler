```python
V0, Ns = 0.0324, 36
a0, a1, a2 = 3.0688, 1.417e-7, 1/(V0*Ns)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v = np.linspace(16, 30, 1000)
i = a0 - a1*np.exp( a2*v -1 )

p = i*v
pMax = p.max()

vMax = v[np.where( p == p.max() )[0][0]]

plt.figure()
ax = plt.subplot(111)
ax.plot(v, i*v)
ax.axvline( vMax )
ax.text( vMax , 49, f'{vMax:.2f}' ) 

ax1 = plt.twinx(ax)
ax1.plot( v, i, color='orange' )
ax1.axhline(0, color='k')
```

$$I = a_0 - a_1 \exp \left( \frac V { N_s V_0 } - 1 \right)$$

Differentiating this w.r.t. $V$ we get:

\begin{align}
\frac {dI} {dV} &= -a_1 \exp \left( \frac {V} {N_sV_o} -1 \right) \cdot \frac {1} {N_sV_o} \\
                &= \left( I - a_0 \right)  \cdot \frac {1} {N_sV_o}
\end{align}

The equation for the power is: $P = IV$

Differentiating this, we get

\begin{align}
\frac {dP} {dV} &= I + V \cdot \frac {dI} {dV} \\ 
                &= I + V \cdot (I - a_o) \cdot \frac 1 {N_s V_0}
\end{align}

Setting this equal to 0, we get the following equation:


\begin{align}
0  &= I + V \cdot (I - a_o) \cdot \frac 1 {N_s V_0}
\end{align}

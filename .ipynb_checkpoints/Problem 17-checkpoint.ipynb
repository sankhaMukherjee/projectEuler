# Problem 17: Number letter counts



If the numbers 1 to 5 are written out in words: one, two, three, four, five, then there are 3 + 3 + 5 + 4 + 4 = 19 letters used in total.

If all the numbers from 1 to 1000 (one thousand) inclusive were written out in words, how many letters would be used?

NOTE: Do not count spaces or hyphens. For example, 342 (three hundred and forty-two) contains 23 letters and 115 (one hundred and fifteen) contains 20 letters. The use of "and" when writing out numbers is in compliance with British usage.


```python
numDict = {
    1: 'one',
    2: 'two',
    3: 'three',
    4: 'four',
    5: 'five',
    6: 'six',
    7: 'seven',
    8: 'eight',
    9: 'nine',
    10: 'ten',
    11: 'eleven',
    12: 'twelve',
    13: 'thirteen',
    14: 'fourteen',
    15: 'fifteen',
    16: 'sixteen',
    17: 'seventeen',
    18: 'eighteen',
    19: 'nineteen',
    20: 'twenty',
    30: 'thirty',
    40: 'forty',
    50: 'fifty',
    60: 'sixty',
    70: 'seventy',
    80: 'eighty',
    90: 'ninety',
    1000: 'one thousand'
}
```

```python
numDict
```

```python
def num(n, numDict):
    
    if n in numDict:
        return numDict[n], numDict
    
    if n >= 100:
        h = n // 100
        t = n % 100
        result = f'{numDict[h]} hundred'
        if t == 0:
            numDict[n] = result
            return result, numDict
        
        result1, numDict1 = num( t, numDict )
        numDict.update( numDict1 )
        result = f'{result} and {result1}'
        numDict[n] = result
        return result, numDict
    
    if n > 20:
        t, u = n//10, n%10
        result = numDict[ t*10 ]
        result = f'{result} {numDict[u]}'
        numDict[n] = result
        return result, numDict
        
```

```python
totals = 0
for i in range(1, 1001):
    result, numDict = num(i, numDict)
    N = len(result.replace(' ', ''))
    totals += N
    print(f'{i:5d} | {N:5d} | {result}')
```
